```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
loan_data=pd.read_csv("loan.csv")
```

```python
loan_data
```

```python
loan_data.isnull().sum()
```

```python
loan_data.isnull().any()

```

```python
new_loan_data=loan_data.replace(to_replace=np.nan,value=0)
```

```python
new_loan_data.isnull().sum()
```

```python
new_loan_data.describe()
```

```python
sns.relplot(x='ApplicantIncome',y='CoapplicantIncome',data=new_loan_data)
```

```python
#most of the applicants income stands between 0 and 20000 and co applicant income is always 1/4th the income of the applicants
```

```python
sns.relplot(x='ApplicantIncome',y='CoapplicantIncome',hue='Credit_History',data=new_loan_data)
```

```python
#if applicant income is more, loan can be approved easily eventhough the coapplicant income is less
#if both applicant and co applicant have higher income loan approval is easy and fast
```

```python
sns.catplot(x='Credit_History',y='LoanAmount',data=new_loan_data)
```

```python
#if loan amount is lesser higher chance of payback, credit history is good for people who took lesser amount as loan
```

```python
sns.jointplot(x='ApplicantIncome',y='LoanAmount',kind='reg',data=new_loan_data)
```

```python
#if applicant income is high, higher chance to get more loan amount.
```
